```python
import pandas as pd
```

```python
df = pd.read_csv('files_/final_clean.csv', index_col= 0, low_memory=False)
```

```python

bins = [-1, 0, 31, 62, 92, 123, 153, 184, 214, 245, 275, 306, 336, 366]
labels = ['Ningún mes', 'Mes 1', 'Mes 2', 'Mes 3', 'Mes 4', 'Mes 5', 'Mes 6', 'Mes 7', 'Mes 8', 'Mes 9', 'Mes 10', 'Mes 11', 'Mes 12']

df['lead_time_group'] = pd.cut(df['lead_time'], bins=bins, labels=labels)
grouped = df[df['country'] == 'DEU'].groupby(['lead_time_group', 'canceled']).size().unstack(fill_value=0)
grouped['total_reservas'] = grouped.sum(axis=1)
grouped['porcentaje_cancelados'] = (grouped['True'] / grouped['total_reservas']) * 100
grouped['country'] = 'DEU'
grouped['lead_period'] = grouped.index
grouped=grouped.drop(columns=['False', 'True'])
DEU = grouped 
```
